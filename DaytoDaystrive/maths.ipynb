```python
import math
```

```python
def countdigit(n):
    return int(math.log10(n)+1)
print(countdigit(1234))
```

```python
n=1234
def revnum(n):
    d=0
    while n>0:
        r=n%10
        d=(d*10)+r
        n//=10
    print(d)
revnum(n)
```

```python
def palindrome(n):
    rev=0
    temp=n
    while n>0:
        d=n%10
        rev=(rev*10)+d
        n//=10
    if temp==rev:
        print("Palindrome")
    else:
        print("Not Palindrome")
palindrome(1001)
```

```python
def gcd(a,b):
    while b!=0:
        a,b=b,a%b
    return a
print(gcd(36,12))
```

```python
def isArmstrong(num):
    k = len(str(num))
    sum = 0
    n = num
    while n > 0:
        ld = n % 10
        sum += ld ** k
        n = n // 10
    return sum == num
print(isArmstrong(153))
```

```python

def findDivisors(n):
    # Initialize an empty
    # list to store the divisors
    divisors = [] 
    # Iterate up to the square
    # root of n to find divisors
    # Calculate the square root of n
    sqrtN = int(math.sqrt(n)) 
    
    # Loop from 1 to the
    # square root of n
    for i in range(1, sqrtN + 1): 
        # Check if i divides n
        # without leaving a remainder
        if n % i == 0: 
            # Add divisor i to the list
            divisors.append(i) 

            # Add the counterpart divisor
            # if it's different from i
            if i != n // i:
                # Add the counterpart
                # divisor to the list
                divisors.append(n // i) 
    
    # Return the list of divisors
    return divisors 
print(findDivisors(12))
```
